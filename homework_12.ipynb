EXERCISE 12.1

Write tests for the Vector class (Homework 6) using the 'unittest' or 'pytest' module. 

```python
from importnb import Notebook
with Notebook():
    import homework_6 as hw6

import unittest
import math

class TestVector(unittest.TestCase):
    def test_equal(self):
        self.assertTrue(hw6.Vector(1,2,3) == hw6.Vector(1,2,3))
        self.assertFalse(hw6.Vector(1,2,3) == hw6.Vector(1,2,4))
    
    def test_not_equal(self):
        self.assertTrue(hw6.Vector(1,2,3) != hw6.Vector(1,2,-3))
        self.assertFalse(hw6.Vector(1,2,3) != hw6.Vector(1,2,3))
    
    def test_adding(self):
        self.assertEqual(hw6.Vector(0,0,1)+hw6.Vector(0,1,0), hw6.Vector(0,1,1))
        
    def test_subtraction(self):
        self.assertEqual(hw6.Vector(1,2,3)-hw6.Vector(0,0,1), hw6.Vector(1,2,2))
        
    def test_dot(self):
        self.assertEqual(hw6.Vector(1,2,3)*hw6.Vector(2,-3,2), 2)
        
    def test_cross(self):
        self.assertEqual(hw6.Vector(1,2,3).cross(hw6.Vector(2,-3,2)),
                        hw6.Vector(13,4,-7))
    
    def test_len(self):
        self.assertEqual(hw6.Vector(1,2,3).length(), math.sqrt(14))
        
suite = unittest.TestLoader().loadTestsFromTestCase(TestVector)
unittest.TextTestRunner().run(suite)
```
